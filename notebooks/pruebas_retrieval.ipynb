```python
from src.embeddings.vector_store import load_vector_store
from src.evaluation.rag_pipeline import answer_with_rag
from src.evaluation.baseline import answer_without_rag

VECTOR_STORE_PATH = "data/embeddings/faiss_index"
vector_store = load_vector_store(VECTOR_STORE_PATH)
```

```python
query = "¿Qué es Retrieval-Augmented Generation?"
retrieved_answer, retrieved_sources = answer_with_rag(query, vector_store)

print("RESPUESTA RAG:\n")
print(retrieved_answer)

print("\nFUENTES:")
for src in retrieved_sources:
    print(f"- {src['source_file']} | página {src['page']}")
```

```python
baseline_answer = answer_without_rag(query)

print("RESPUESTA BASELINE:\n")
print(baseline_answer)
```

```python
questions = [
    "¿Qué es Retrieval-Augmented Generation?",
    "¿Para qué sirven los embeddings en un sistema RAG?",
    "¿Qué diferencia hay entre un sistema RAG textual y uno multimodal?",
    "¿Por qué el contenido visual puede ser problemático en documentos académicos?",
    "¿Qué papel tiene una base de datos vectorial en este sistema?"
]
```

```python
for q in questions:
    print("=" * 100)
    print("PREGUNTA:", q)

    rag_answer, rag_sources = answer_with_rag(q, vector_store)
    baseline_answer = answer_without_rag(q)

    print("\n--- RAG ---")
    print(rag_answer)

    print("\n--- FUENTES RAG ---")
    for src in rag_sources:
        print(f"- {src['source_file']} | página {src['page']}")

    print("\n--- BASELINE ---")
    print(baseline_answer)
    print("=" * 100)
```

```python
results = []

for q in questions:
    rag_answer, rag_sources = answer_with_rag(q, vector_store)
    baseline_answer = answer_without_rag(q)

    results.append({
        "question": q,
        "rag_answer": rag_answer,
        "rag_sources": rag_sources,
        "baseline_answer": baseline_answer
    })

len(results)
```

```python
import json
from pathlib import Path

Path("experiments").mkdir(exist_ok=True)

with open("experiments/notebook_comparison_results.json", "w", encoding="utf-8") as f:
    json.dump(results, f, indent=2, ensure_ascii=False)

print("Resultados guardados.")
```
